```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
df_sales =  pd.read_csv('sample_sales_data.csv')
df_sales
```

```python
df_sales.isna().sum()
Num_DF = df_sales.select_dtypes(include='number')
```

```python
Num_DF
```

```python
sns.heatmap(Num_DF.corr())
```

```python
sns.scatterplot(data=Num_DF, x='Price', y='TotalAmount') \
    .set(title="Scatter Plot: Quantity vs Total Amount")
```

```python
sns.scatterplot(data=Num_DF, x='Quantity', y='TotalAmount') \
    .set(title="Scatter Plot: Quantity vs Total Amount")
```

```python
sns.scatterplot(data=Num_DF, x='Quantity', y='Price') \
    .set(title="Scatter Plot: Quantity vs Total Amount")
```

```python
sns.barplot(data=df_sales, x='Category', y='TotalAmount', estimator=sum,ci=None)
```

```python
sns.boxplot(data=df_sales, x='Category', y='TotalAmount')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Plot the correlation heatmap on the first subplot
sns.heatmap(Num_DF, cmap='coolwarm', ax=axes[0])
axes[0].set_title('Correlation Matrix')

# Plot scatter plots on the second subplot
sns.scatterplot(data=Num_DF, x='Quantity', y='TotalAmount', ax=axes[1], color='blue', label='Quantity vs TotalAmount')
sns.scatterplot(data=Num_DF, x='Price', y='TotalAmount', ax=axes[1], color='red', label='Price vs TotalAmount')

axes[1].set_title('Scatter Plots')
axes[1].legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
