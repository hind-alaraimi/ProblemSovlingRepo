```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

# Function for One-Sample t-Test
def one_sample_t_test(sample_mean, population_mean, sample_sd, sample_size, alpha=0.01):
    # Calculate the t-value
    t_value = (sample_mean - population_mean) / (sample_sd / np.sqrt(sample_size))

    # Degrees of freedom
    df = sample_size - 1

    # Calculate the critical t-value (one-tailed, left)
    t_critical = stats.t.ppf(alpha, df)

    # Calculate the p-value
    p_value = stats.t.cdf(t_value, df)

    # Decision
    if t_value < t_critical:
        decision = "Reject the null hypothesis (H0)."
    else:
        decision = "Fail to reject the null hypothesis (H0)."

    # Displaying results
    print(f"t-value: {t_value:.4f}")
    print(f"Critical t-value: {t_critical:.4f}")
    print(f"p-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    # Plotting the t-distribution
    x = np.linspace(-4, 4, 500)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, color='royalblue')
    plt.fill_between(x, y, where=(x <= t_critical), color='red', alpha=0.4, label='Rejection Region')
    plt.axvline(t_value, color='green', linestyle='--', label=f't-value = {t_value:.4f}')
    plt.title("t-Distribution with Rejection Region")
    plt.xlabel("t-value")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

# Example Usage
if __name__ == "__main__":
    # Given values
    sample_mean = 1.8
    population_mean = 2
    sample_sd = 0.15
    sample_size = 10

    # Running the t-test
    one_sample_t_test(sample_mean, population_mean, sample_sd, sample_size)
```
