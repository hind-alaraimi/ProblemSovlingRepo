```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Setting seed for reproducibility
np.random.seed(42)

# Generating data for different distributions
no_skew = np.random.normal(loc=0, scale=1, size=1000)
positive_skew = np.random.exponential(scale=1, size=1000)
negative_skew = -np.random.exponential(scale=1, size=1000)

def plot_skewness_types():
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('Skewness Types: No Skew, Positive Skew, Negative Skew')

    distributions = {
        'No Skew (Normal)': no_skew,
        'Positive Skew': positive_skew,
        'Negative Skew': negative_skew
    }

    for ax, (name, dist_data) in zip(axes, distributions.items()):
        x_values = np.linspace(min(dist_data), max(dist_data), 1000)
        ax.plot(x_values, stats.gaussian_kde(dist_data)(x_values), color='black')

        mean = np.mean(dist_data)
        median = np.median(dist_data)
        mode = x_values[np.argmax(stats.gaussian_kde(dist_data)(x_values))]

        ax.axvline(mean, color='green', linestyle='--', label='Mean')
        ax.axvline(median, color='orange', linestyle='--', label='Median')
        ax.axvline(mode, color='blue', linestyle='--', label='Mode')

        ax.set_title(name)
        ax.legend()
        ax.set_xlabel('Values')
        ax.set_ylabel('Density')

    plt.tight_layout()
    plt.show()

# Running the plot function
if __name__ == '__main__':
    plot_skewness_types()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Setting seed for reproducibility
np.random.seed(42)

# Generate data for different distributions
normal_dist = np.random.normal(loc=0, scale=1, size=1000)  # Normal Distribution (Mesokurtic)
positive_skew = np.random.exponential(scale=1, size=1000)  # Positive Skew
negative_skew = -np.random.exponential(scale=1, size=1000)  # Negative Skew

# Leptokurtic (high kurtosis, peak-shaped)
leptokurtic = np.random.normal(loc=0, scale=1, size=1000)
leptokurtic = np.concatenate([leptokurtic, np.random.normal(loc=0, scale=0.2, size=100)])  # Adding outliers

# Platykurtic (low kurtosis, flat distribution)
platykurtic = np.random.uniform(low=-3, high=3, size=1000)

# Mesokurtic (normal distribution with kurtosis around 3)
mesokurtic = np.random.normal(loc=0, scale=1, size=1000)  # Same as normal, mesokurtic by nature

# Calculate skewness and kurtosis
distributions = {
    'Normal': normal_dist,
    'Positive Skew': positive_skew,
    'Negative Skew': negative_skew,
    'Leptokurtic': leptokurtic,
    'Platykurtic': platykurtic,
    'Mesokurtic': mesokurtic
}

skew_kurtosis_values = {}
for name, dist_data in distributions.items():
    skew_kurtosis_values[name] = {
        'Skewness': stats.skew(dist_data),
        'Kurtosis': stats.kurtosis(dist_data)
    }

# Plotting different distributions
def plot_distributions():
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle('Skewness and Kurtosis of Different Distributions')

    # Distribution labels
    dist_labels = list(distributions.keys())

    # Loop over distributions and plot them
    for ax, dist_label in zip(axes.flatten(), dist_labels):
        dist_data = distributions[dist_label]

        # Create a Kernel Density Estimate (KDE) plot
        x_values = np.linspace(min(dist_data), max(dist_data), 1000)
        ax.plot(x_values, stats.gaussian_kde(dist_data)(x_values), color='black')

        # Plot mean, median, and mode
        mean = np.mean(dist_data)
        median = np.median(dist_data)
        mode = x_values[np.argmax(stats.gaussian_kde(dist_data)(x_values))]

        ax.axvline(mean, color='green', linestyle='--', label='Mean')
        ax.axvline(median, color='orange', linestyle='--', label='Median')
        ax.axvline(mode, color='blue', linestyle='--', label='Mode')

        # Set title and labels
        ax.set_title(f'{dist_label} (Skew: {skew_kurtosis_values[dist_label]["Skewness"]:.2f}, '
                     f'Kurt: {skew_kurtosis_values[dist_label]["Kurtosis"]:.2f})')
        ax.set_xlabel('Values')
        ax.set_ylabel('Density')
        ax.legend()

    plt.tight_layout()
    plt.show()

# Running the plot function
plot_distributions()
```
